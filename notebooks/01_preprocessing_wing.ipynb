```python
df_clean = df.drop_duplicates().reset_index(drop=True)
df_clean.shape
```

```python
import numpy as np
from scipy import stats

# Convert to numpy array
X = df_clean.values

# Compute z-scores
z = np.abs(stats.zscore(X))

# Keep rows where all features have |z| < 3
mask = (z < 3).all(axis=1)

# Apply mask
X_filtered = X[mask]
df_filtered = df_clean.iloc[mask].reset_index(drop=True)

X.shape, X_filtered.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_filtered)

X_scaled[:5]
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
X_pca_2d = pca.fit_transform(X_scaled)

X_pca_2d[:5]
```

```python
pca.explained_variance_ratio_
```

```python
pca.explained_variance_ratio_.sum()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 5))
plt.scatter(X_pca_2d[:, 0], X_pca_2d[:, 1], s=5)

plt.title("PCA (2D)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
from sklearn.cluster import KMeans

# try 4 clusters just as an example
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans_labels = kmeans.fit_predict(X_pca_2d)

# check first 10 labels
kmeans_labels[:10]
```

```python
plt.figure(figsize=(6,5))
plt.scatter(
    X_pca_2d[:, 0],
    X_pca_2d[:, 1],
    c=kmeans_labels,
    s=5
)
plt.title("KMeans (k=4) on PCA (2D)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
from sklearn.metrics import silhouette_score

sil_score = silhouette_score(X_pca_2d, kmeans_labels)
sil_score
```

```python
import pandas as pd

df_pca = pd.DataFrame({
    "PC1": X_pca_2d[:, 0],
    "PC2": X_pca_2d[:, 1]
})

df_pca.to_csv("../data/pca_2d_results.csv", index=False)
```

```python
import umap.umap_ as umap

umap_2d = umap.UMAP(
    n_components=2,
    n_neighbors=15,
    min_dist=0.1,
    random_state=42
)

X_umap_2d = umap_2d.fit_transform(X_scaled)
X_umap_2d[:5]
```

```python
plt.figure(figsize=(6,5))
plt.scatter(X_umap_2d[:,0], X_umap_2d[:,1], s=5)
plt.title("UMAP (2D)")
plt.xlabel("UMAP1")
plt.ylabel("UMAP2")
plt.show()
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.2, min_samples=20)  # eps may need tweaking
dbscan_labels = dbscan.fit_predict(X_umap_2d)

# unique labels (clusters)
import numpy as np
np.unique(dbscan_labels)
```

```python
plt.figure(figsize=(6,5))
plt.scatter(
    X_umap_2d[:, 0],
    X_umap_2d[:, 1],
    c=dbscan_labels,
    s=5
)
plt.title("DBSCAN on UMAP (2D)")
plt.xlabel("UMAP1")
plt.ylabel("UMAP2")
plt.show()
```

```python
valid_mask = dbscan_labels != -1
if len(np.unique(dbscan_labels[valid_mask])) > 1:
    sil_dbscan = silhouette_score(X_umap_2d[valid_mask], dbscan_labels[valid_mask])
    sil_dbscan
else:
    sil_dbscan = None
sil_dbscan
```
